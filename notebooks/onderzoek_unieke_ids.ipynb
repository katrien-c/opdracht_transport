```python
import pandas as pd
import os
import json
```

```python


list = []
requests_data_folder = "../data/requests/"
for folder_name in os.listdir(requests_data_folder):
    folder_path = os.path.join(requests_data_folder, folder_name)
    if os.path.isdir(folder_path):
        depotroute = folder_name.split("-", 1)[0]
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            if os.path.isfile(file_path):
                with open(file_path, "r", encoding="utf-8") as f:
                    data = json.load(f)
                    for task in data['tasks']:
                        id = task['id']
                        latitude = task['address']['latitude']
                        longitude = task['address']['longitude']
                        depotroute = depotroute
                        list.append((id, latitude, longitude, depotroute))

df_data = pd.DataFrame(list, columns=['id', 'latitude', 'longitude', 'depotroute']).sort_values(by=['id', 'latitude', 'longitude', 'depotroute']).drop_duplicates().copy()

df_data.to_csv('df_data.csv', index=False)
# toon het aantal rijen van de dataframe
print(f"Aantal unieke rijen: {len(df_data)}")

# drop de twee laatste kolommen en verwijder de duplicaten en tel dan het aantal rijen, 
# als het hetzelfde is als hierboven, dan zijn de ids uniek
df_data_ids_only = df_data.drop(columns=['latitude', 'longitude']).drop_duplicates()

print(f"Aantal unieke ids: {len(df_data_ids_only)}")

if (len(df_data) == len(df_data_ids_only)):
    print("De ids zijn uniek")
else:
    print("De ids zijn niet uniek")



```
